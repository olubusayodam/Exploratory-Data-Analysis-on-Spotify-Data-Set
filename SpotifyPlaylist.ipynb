```python
#Which artist has the highest average popularity for their songs?
```

```python
avg_popularity_by_artist = df.groupby('artists')['popularity'].mean()
max_avg_popularity_artist = avg_popularity_by_artist.idxmax()
print(f"Artist with Highest Average Popularity: {max_avg_popularity_artist}")
```

```python
#What is the average duration of songs in the dataset?
```

```python
avg_duration = df['duration_ms'].mean()
print(f"Average Duration: {avg_duration} ms")
```

```python
#How many unique artists are present in the dataset?
```

```python
unique_artists_count = df['artists'].explode().nunique()
print(f"Number of Unique Artists: {unique_artists_count}")
```

```python
#What is the distribution of speechiness in the dataset?
```

```python
plt.figure(figsize=(8, 6))
sb.histplot(df['speechiness'], bins=20, kde=True)
plt.title('Distribution of Speechiness')
plt.xlabel('Speechiness')
plt.ylabel('Frequency')
plt.show()
```

```python
#How many songs were released in each decade (assuming 10-year intervals)?
```

```python
df['release_decade'] = (pd.to_datetime(df['release_date'], errors='coerce')
                        .dt.year // 10 * 10)
songs_per_decade = df['release_decade'].value_counts().sort_index()
print("Songs Released per Decade:\n", songs_per_decade)
```

```python
#What is the average danceability of songs in major mode compared to minor mode?
```

```python
avg_danceability_major = df[df['mode'] == 1]['danceability'].mean()
avg_danceability_minor = df[df['mode'] == 0]['danceability'].mean()
print(f"Average Danceability (Major Mode): {avg_danceability_major}")
print(f"Average Danceability (Minor Mode): {avg_danceability_minor}")
```

```python
#How many songs were released in each month?
```

```python
df['release_month'] = pd.to_datetime(df['release_date'], errors='coerce').dt.month
songs_per_month = df['release_month'].value_counts().sort_index()
print("Songs Released per Month:\n", songs_per_month)
```

```python
#How many songs were released per year
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Extracting the release year
df['release_year'] = df['release_date'].dt.year

# Counting the number of songs per year
songs_per_year = df['release_year'].value_counts().sort_index()

# Creating a DataFrame for better tabular representation
songs_per_year_df = pd.DataFrame({'Year': songs_per_year.index, 'Number of Songs': songs_per_year.values})

# Displaying the table
print("Number of Songs Released Each Year:")
print(songs_per_year_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataset is stored in a DataFrame named 'df'
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Extracting the release year
df['release_year'] = df['release_date'].dt.year

# Counting the number of songs per year
songs_per_year = df['release_year'].value_counts().sort_index()

# Plotting the results
plt.figure(figsize=(12, 6))
songs_per_year.plot(marker='o')
plt.title('Number of Songs Released Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.show()
```
